```python
import numpy
import pandas
import matplotlib.pyplot as mp
df = pandas.read_csv('winequality-white.csv', delimiter=';')
df = df.loc[df['residual sugar']<1]
df = df[['density', 'pH']]
df = df.iloc[df.index%2==0]
df['pH'].hist()
mp.title('pH')
mp.show()
df['density'].plot()
mp.title('density')
mp.show()
```
